```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2 * x + 5 + np.random.normal(0, 2, len(x)).reshape(-1, 1)
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
def least_sqaure_method(x, y):
    x = np.hstack((np.ones((100, 1)), x))
    return np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
```

```python
theta = least_sqaure_method(x, y)
theta
```

```python
plt.figure()
plt.plot(x, y)
y_hat = np.hstack((np.ones((100, 1)), x)).dot(theta)
plt.plot(x, y_hat)
plt.text((x.max() - x.min())/2+0.2, (y_hat.max()-y_hat.min())/2-0.5, f'y={theta[0,0]:.2f}x+{theta[1,0]:.2f}')
plt.show()
```

```python
def multFit(x, y ,n = 3):
    x_bar = np.hstack((np.ones(x.shape), x))
    if n >= 2:
        for i in range(2, n + 1):
            x_bar = np.hstack((x_bar, x**i))
    print(x_bar.shape)
    return x_bar, np.linalg.inv(x_bar.T.dot(x_bar)).dot(x_bar.T).dot(y)
```

```python
x_bar, theta_1 = multFit(x, y, 5)
y_bar = x_bar.dot(theta_1)
plt.figure()
plt.plot(x, y)
plt.plot(x, y_bar)
plt.show()
```
